```python
df_financials = pd.read_excel("data/movies_db.xlsx", sheet_name= "financials")
df_financials.head()
```

```python
def std_curr(x):
    if x == '$$' or x == 'Dollars':
        return "USD"
    return x 

df_financials = pd.read_excel("data//movies_db.xlsx", sheet_name= "financials",
                              
                              converters= {'currency': std_curr})
df_financials.head()
```

```python
df_merged = pd.merge(df_movies, df_financials, on= "movie_id")
```

```python
df_merged.to_excel("movies_merged.xlsx", sheet_name= "merged movies", index=False)
```
